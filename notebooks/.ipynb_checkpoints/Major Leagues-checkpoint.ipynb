### Feature Engineering

```python
# Create new column for storing the match outcome
```

```python
#Return 2 if home team wins
#Return 1 if match ties
#Return 0 if away team wins
def label(data):
    if data["score1"] > data["score2"]:
        return 2
    elif data["score2"] > data["score1"]:
        return 0
    elif data["score2"] == data["score1"]:
        return 1
```

```python
games["winner"] = games.apply(lambda games:label(games),axis=1)
```

```python
plt.figure(figsize=(8,8))
labels = 'HOME TEAM', 'AWAY TEAM', 'DRAW'
colors = ['yellowgreen', 'lightcoral', 'lightskyblue']

games["winner"].value_counts().plot.pie(autopct = "%1.0f%%",labels=labels,
                                             colors =colors)
my_circ = plt.Circle((0,0),.7,color = "white")
plt.gca().add_artist(my_circ)
plt.title("Proportion of Game Outcomes")
plt.show()
```

```python
#Create another data frame object with the features used for predicting the scores
games_filtered = games[['team1', 'team2','elo1','elo2', 'winner']].copy()
```

```python
games_filtered.head()
```

```python
#Use one hot encoding to convert symbolic data to numerical data
final=pd.get_dummies(games_filtered,prefix=['home_team','away_team'],columns=['team1','team2'])
```

```python
final.head()
```

### Regression Modeling

#### Predicting using both score and elo rating features

```python
#Define features
X=final.drop(['winner'], axis=1)
```

```python
#Define target
y=final['winner']
```

```python
y=y.astype('int')
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 12)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
prediction = dict()
```

```python
prediction['Logistic'] = model.predict(x_test)
```

```python
print('Accuracy: ',accuracy_score(y_test, prediction['Logistic']))
```

```python
conf_mat_logist = confusion_matrix(y_test, prediction['Logistic'])
```

```python
print('Logist \r', conf_mat_logist)
```

```python
#Print first 20 predictions
print(prediction['Logistic'][0:20])
```

### Predict actual scores for home and away team

```python
games['teamA'] = pd.factorize(games.team1)[0] + 1
```

```python
games['teamB'] = pd.factorize(games.team2)[0] + 1
```

```python
games.head()
```

```python
games = games.drop(columns=['date','season','team1','team2','result1','neutral','playoff'])
```

```python
final = games.copy()
```

```python
#Define features
x=final.drop(['score1','score2'], axis=1)

#Define target
y_home=final['score1']
y_away=final['score2']
```

```python
x_home_train,x_home_test,y_home_train,y_home_test=train_test_split(x,y_home,test_size=0.2,random_state=0)
x_away_train,x_away_test,y_away_train,y_away_test=train_test_split(x,y_away,test_size=0.2,random_state=0)
k_fold = KFold(n_splits=5, shuffle=True, random_state=0)
```

```python
param_grid = dict(C=(0.0001,0.001,0.005,0.01,0.1,0.5,1))
```

```python
homelog_reg1 = GridSearchCV(LogisticRegression(penalty="l1"),param_grid=param_grid,scoring="f1_macro")
```

```python
homelog_reg1.fit(x_home_train,y_home_train)
print(homelog_reg1.best_params_)
```

### The columns in the confusion matrix below are representing the predicted game score for each home team, and index is representing the actual game score

```python
cm=confusion_matrix(y_home_train,homelog_reg1.predict(x_home_train))
cm=pd.DataFrame(cm)
cm.T
```

```python
awaylog_reg1 = GridSearchCV(LogisticRegression(penalty="l1"),param_grid=param_grid,scoring="f1_macro")
```

```python
awaylog_reg1.fit(x_away_train,y_away_train)
print(awaylog_reg1.best_params_)
```

### The columns in the confusion matrix below are representing the predicted game score for each away team, and index is representing the actual game score

```python
cm=confusion_matrix(y_home_train,awaylog_reg1.predict(x_home_train))
cm=pd.DataFrame(cm)
cm.T
```
